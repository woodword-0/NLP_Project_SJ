```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
text = "If you were the NWS wth a rotating storm w/ a report of a 'HUGE' / 'MASSIVE' / 'VIOLENT' tornado what would you do? https://t.co/J3dI85IST5"
```

```python
text
```

```python
import string
```

```python
string.punctuation
```

```python
text_p = "".join([char for char in text if char not in string.punctuation])
```

```python
text = text.lower()
```

```python
text
```

```python
import re
```

```python
text_dn = re.sub(r'https?:\/\/.*[\r\n]*','',text ) #remove https and replace with blank text
```

```python
text_dn
```

```python
text_p = "".join([char for char in text_dn if char not in string.punctuation]) #remove punctuation
```

```python
text_p
```

```python
import nltk
nltk.download('stopwords')
```

```python
#stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
print(stop_words)
```

```python
#tokenize
from nltk import word_tokenize
from nltk import downloader
nltk.download('punkt')
```

```python
words = word_tokenize(text_p)
```

```python
words
```

```python
filtered_words = [word for word in words if word not in stop_words] #take out any word that is not a stop word
```

```python
filtered_words
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()
```

```python
stemmed_words = [porter.stem(word) for word in filtered_words]
print(stemmed_words)
```
